```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0.01,1.01,0.01)

def b(x):
    x2= np.exp(x)
    return x2
```

```python
y = b(x)
fig, ax = plt.subplots(figsize=(7,7))

ax.set_xlabel('Time[milliseconds]')
ax.set_ylabel('Awsomeness')
ax.plot(x,y)

plt.savefig('persona7.png',bbox_inches='tight',facecolor='white')
```
