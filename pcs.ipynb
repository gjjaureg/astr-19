```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0.01,1.01,0.01)

def s(x):
    x2 = np.sin(x)
    return x2
def c(x):
    x3 = np.cos(x)
    return x3
```

```python
f, axarr = plt.subplots(1,2)

axarr[0].plot(s(x),x)
axarr[0].set_xlabel('sin(x)')
axarr[0].set_ylabel('x')
axarr[0].set_title(r'$\sin(x)$')

axarr[1].plot(c(x),x)
axarr[1].set_xlabel('cos(x)')
axarr[1].set_ylabel('x')
axarr[1].set_title(r'$\cos(x)$')


```
