```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.default_rng()
rng2 = rng.choice(1000, 1000)

plt.hist(rng2, alpha=0.5, bins=100)
plt.xlabel('100 bins (x-axis)')
plt.ylabel('1000 random numbers from random choice (y-axis)')
plt.savefig('prompt9', bbox_inches='tight',facecolor='white')
```
